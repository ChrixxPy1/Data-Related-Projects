```python
#Which year has the highest release of movies?
data = df.groupby('release_year').count()['id']
#data
#make group of the data according to their release year and count the total number of movies in each year and pot.
data = df.groupby('release_year').count()['id'].plot(xticks = np.arange(1960,2016,5))

#set the figure size and labels
sns.set(rc={'figure.figsize':(10,5)})
plt.title("Year Vs Number Of Movies",fontsize = 14)
plt.xlabel('Release year',fontsize = 13)
plt.ylabel('Number Of Movies',fontsize = 13)
#set the style sheet
sns.set_style("whitegrid")
```

```python
#Which Movie Has The Highest Or Lowest Profit? Top 10 movies which earn highest profit?
df['Movie_Profit'] = df['revenue'] - df['budget']
def find_minmax(x):
    #use the function 'idmin' to find the index of lowest profit movie.
    min_index = df[x].idxmin()
    
    #use the function 'idmax' to find the index of Highest profit movie.
    high_index = df[x].idxmax()
    
    high = pd.DataFrame(df.loc[high_index,:])
    low = pd.DataFrame(df.loc[min_index,:])
    
    #print the movie with high and low profit
    print("Movie Which Has Highest "+ x + " : ",df['original_title'][high_index])
    print("Movie Which Has Lowest "+ x + "  : ",df['original_title'][min_index])
    return pd.concat([high,low],axis = 1)

#call the find_minmax function.
find_minmax('Movie_Profit')
```

```python
#make a plot which contain top 10 movies which earn highest profit.
#sort the 'Profit' column in decending order and store it in the new dataframe,
wer = pd.DataFrame(df['Movie_Profit'].sort_values(ascending = False))
wer['original_title'] = df['original_title']
data = list(map(str,(wer['original_title'])))
x = list(data[:10])
y = list(wer['Movie_Profit'][:10])

#make a plot usinf pointplot for top 10 profitable movies.
ax = sns.pointplot(x=y,y=x)

#setup the figure size
sns.set(rc={'figure.figsize':(10,5)})
#setup the title and labels of the plot.
ax.set_title("Top 10 Profitable Movies",fontsize = 15)
ax.set_xlabel("Profit",fontsize = 13)
sns.set_style("darkgrid")
```

```python
#find the movie with highest and lowest budget.
#use 'find_minmax' function which you made earlier.
#make sure that movie with budget zero didn't affect the result.
#so change the zero into NAN in budget column.
df['budget'] = df['budget'].replace(0,np.NAN)
find_minmax('budget')
```

```python
#make a plot which contain top 10 highest budget movies.
#sort the 'budget' column in decending order first, and store it in the new dataframe that you'll create.
new_sort = pd.DataFrame(df['budget'].sort_values(ascending = False))
new_sort['original_title'] = df['original_title']
data = list(map(str,(new_sort['original_title'])))

#extract the top 10 budget movies data from the list and dataframe.
x = list(data[:10])
y = list(new_sort['budget'][:10])

#plot the figure and setup the title and labels.
ax = sns.pointplot(x=y,y=x)
sns.set(rc={'figure.figsize':(10,5)})
ax.set_title("Top 10 High Budget Movies",fontsize = 15)
ax.set_xlabel("Budget",fontsize = 13)
sns.set_style("darkgrid")
```

```python
#top 10 movies which made highest revenue.
#sort the 'revenue' column in decending order and store it in the new dataframe.
top_rev = pd.DataFrame(df['revenue'].sort_values(ascending = False))
top_rev['original_title'] = df['original_title']
dat = list(map(str,(top_rev["original_title"])))

#now extract the top 10 movies which madethe highst revenue
x = list(dat[:10])
y = list (top_rev['revenue'][:10])

#you can now plot the figure 
axs = sns.pointplot(x=y, y=x)
sns.set(rc = {'figure.figsize' : (12, 7)})
axs.set_title("Top 10 Movies With Highest Revenue")
axs.set_xlabel("Revenue", fontsize = 15)
sns.set_style("darkgrid")

```

```python
#movie with the longest and shortest runtime
info = pd.DataFrame(df['runtime'].sort_values(ascending = False))
info['original_title'] = df['original_title']
data = list(map(str,(info['original_title'])))

#extract the top 10 longest duraton movies data from the list and dataframe.
x = list(data[:10])
y = list(info['runtime'][:10])

#make the point plot and setup the title and labels.
ax = sns.pointplot(x=y,y=x)
sns.set(rc={'figure.figsize':(10,5)})
ax.set_title("Top 10 Longest Movies",fontsize = 15)
ax.set_xlabel("Runtime",fontsize = 13)
sns.set_style("darkgrid")
```

```python
#how does revenue change according to their budget.
#make a scatter plot using 'regplot' between ''revenue' and 'budget'.

ad = sns.regplot(x = df['revenue'], y = df['budget']);
ad.set_title('Revenue vs Budget', fontsize = 15)
ad.set_xlabel('Revenue', fontsize = 15)
ad.set_ylabel('Budget', fontsize = 15)

sns.set_style('darkgrid')



#find the correlation between them
#convert zero into NAN in budget and revenue column for the exact calculation.


df['revenue'] = df['revenue'].replace(0, np.NAN)
df['Movie_Profit'] = df['Movie_Profit'].replace(0, np.NAN)
df['budget'] = df['budget'].replace(0, np.NAN)
daa_corr = df.corr()
print('The correlation between Revenue and Budget is : ', daa_corr.loc['revenue', 'budget'])
```

```python
#how does vote average(ratings) change according to the release year.
#make a scatter plot using 'regplot' between 'year' and 'vote average'.

af = sns.regplot(x = df['release_year'], y = df['vote_average']);
af.set_title('Release Year vs Vote Average', fontsize= 15)
af.set_xlabel('Vote Average', fontsize = 15)
af.set_ylabel('Release Year', fontsize = 15)
```

```python
#how does vote average(ratings) change according to the release year.

af = sns.regplot(x=df['popularity'],y=df['Movie_Profit'],color='c')

#setup the title and the labels of the scatter plot.
af.set_title("Popularity Vs Profit",fontsize=13)
af.set_xlabel("Popularity",fontsize=12)
af.set_ylabel("Profit",fontsize=12)

#setup the figure size.
sns.set(rc={'figure.figsize':(6,4)})
sns.set_style("whitegrid")

#find the correlation between them.
print("Correlation Between Popularity And Profit : ",daa_corr.loc['popularity','Movie_Profit'])
```
